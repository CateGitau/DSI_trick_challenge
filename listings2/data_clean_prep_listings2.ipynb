# Data Cleaning

```python
import pandas as pd
import numpy as np
```

```python
# import csv
data = pd.read_csv("/home/cate/Cate/DSI/DSI_trick_challenge/final.csv",header = 0)
```

```python
data.head()
```

```python
data = data.astype(object)
```

```python
cols = data.columns
cols
```

```python
datas = data.unstack().str.strip('[]').str.split(', ', expand=True)
```

```python
datas
```

```python
un_data = datas.unstack()
```

```python
df = un_data.reset_index()
```

```python
new_data = df.T
```

```python
new_data.columns = new_data.iloc[0]
new_data
```

```python
new_data.reset_index(drop=True, inplace=True)
```

```python
final = new_data.iloc[1:]
```

```python
final.shape
```

```python
final.dtypes
```

```python
final['bathrooms'] = final['bathrooms'].apply(lambda x: str(x).replace("'", "").strip())
```

```python
final['bedrooms'] = final['bedrooms'].apply(lambda x: str(x).replace("'", "").strip())
final['parking'] = final['parking'].apply(lambda x: str(x).replace("'", "").strip())
final['location'] = final['location'].apply(lambda x: str(x).replace("'", "").strip())
final['floor_size'] = final['floor_size'].apply(lambda x: str(x).replace("m²", "").strip())
```

```python
final['floor_size'] = final['floor_size'].apply(lambda x: str(x).replace("'", "").strip())
```

```python
final['price'] = final['price'].apply(lambda x: str(x).replace("xa0", "").strip())
```

```python
final['price'] = final['price'].apply(lambda x: x.replace("R", "").strip())
final
```

```python
import re
final['price'] = final['price'].apply(lambda x: re.sub("[^0-9]", "",x))
```

```python
final
```

```python
final.to_csv("/home/cate/Cate/DSI/DSI_trick_challenge/final_cleaner.csv", index = False)
```
